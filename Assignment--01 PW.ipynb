2. What is an Interpreter in Python?
  - An interpreter in Python is like having a virtual assistant for your code. It reads
and executes your commands line by line, which is super handy for catching
errors quickly. With an interpreter, you can see the output of each line
immediately, making it easier to understand and debug your code. This real-time
feedback loop not only speeds up the development process but also encourages
experimentation and learning. So, if you're looking to test out short snippets of
code or play around with new ideas, Python's interactive nature makes it a
perfect choice.


3. What Are Pre-defined Keywords in Python ?
   - Python has certain reserved words, called keywords, that have special
meanings and can’t be used as variable names. Some of these include: - `if` -
`for` - `while` - `import` - `def` - `class` - `True` - `False` - `None` These keywords
are essential in shaping how the language works


4. Can Keywords Be Used as Variable Names ?
   - No, you can’t use keywords as variable names in Python. Since they are
reserved for specific functions in the language, using them as names would
cause an error


5. What is Mutability in Python ?
   - Mutability refers to whether you can change an object after it’s created. Some
objects are mutable, meaning their content can be changed, while others are
immutable, meaning they can’t be changed once they’re created


6. Why Are Lists Mutable, but Tuples Are Immutable ?
    - Lists are mutable, meaning you can change, add, or remove items after
creation. - **Tuples** are immutable, meaning once you create a tuple, you can't
modify it. This makes tuples more reliable and secure, and that’s why you can
use them as keys in dictionaries


7. What is the Difference Between “==” and “is” Operators in Python?
   - checks if the **values** of two objects are the same. - `is` checks if two objects
are actually the **same** object in memory—essentially, if they are identical, not
just equal


8. What Are Logical Operators in Python?
   - Logical operators are used to combine multiple conditions: - **`and`**: Returns
`True` if both conditions are true. - **`or`**: Returns `True` if at least one condition
is true. - **`not`**: Reverses the truth value of a condition (makes `True` into
`False`, and vice versa)  


9. What is Type Casting in Python?
    - Type casting is when you convert a value from one data type to another. You
can do this **implicitly**, where Python does it automatically, or **explicitly**,
where you do it manually using functions like `int()` or `float()


10. What is the Difference Between Implicit and Explicit Type Casting?
   - Implicit casting** is when Python automatically changes a smaller data type
into a larger one, like converting an `int` to a `float`. - **Explicit casting** is when
you manually convert one type to another using specific functions (like `int()` or
`float()`)


11. What is the Purpose of Conditional Statements in Python?
   - Conditional statements (`if`, `elif`, `else`) let you control what the program does
based on whether a condition is true or false. They help you make decisions.


12. How Does the `elif` Statement Work?

    - The `elif` (short for "else if") statement lets you check multiple conditions in
sequence. If the first condition (`if`) is false, Python checks the conditions in the
`elif` block. This way, you can test several possibilities in a clean and organized
way


13. What is the Difference Between `for` and `while` Loops?

  - A **`for` loop** is used when you want to loop through a fixed set of items (like
a list or range) A **`while` loop** keeps going as long as a condition is true. You
use it when the number of iterations is uncertain or depends on something that
changes during execution


14. Describe a Scenario Where a `while` Loop is More Suitable Than a `for`
Loop.
  - A `while` loop is perfect when you don’t know how many times the loop will run
ahead of time, or when the condition to end the loop might change based on user

```python
# 1. Simple program to print "Hello, World!"
print("Hello, World!")

# 2. A program that tells your name and age.
name = "laxman Rajput";
age = 22
print(f"My name is {name} and I am {age} years old.")

# 3. Code to print all the keywords in Python.
import keyword
print("Python's reserved keywords are:", keyword.kwlist)

# 4. Check if a specific word is a Python keyword.
word = "def"
if keyword.iskeyword(word):
    print(f"{word} is a Python keyword.")
else:
    print(f"{word} is not a Python keyword.")

# 5. Function to show the difference between mutable and immutable types.
def demonstrate_mutability(lst, tup):
    lst[0] = 100
    try:
        tup[0] = 100
    except TypeError:
        print("Cannot modify a tuple!")

my_list = [1, 2, 3]
my_tuple = (1, 2, 3)
demonstrate_mutability(my_list, my_tuple)

# 6. Perform basic arithmetic on two numbers entered by the user.
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
print("Addition:", num1 + num2)
print("Subtraction:", num1 - num2)
print("Multiplication:", num1 * num2)
print("Division:", num1 / num2)

# 7. Demonstrate how logical operators work in Python.
x = True
y = False
print("x and y:", x and y)
print("x or y:", x or y)
print("not x:", not x)

# 8. Lists and tuples: Demonstrate how changing an element works differently.
my_list = [1, 2, 3]
my_tuple = (1, 2, 3)
my_list[0] = 10
print("Modified list:", my_list)
try:
    my_tuple[0] = 10
except TypeError:
    print("Cannot modify a tuple!")

# 9. Convert a user's input to integer, float, and boolean types.
user_input = input("Enter something: ")
try:
    int_input = int(user_input)
    print(f"Converted to integer: {int_input}")
except ValueError:
    print("Cannot convert to integer.")

try:
    float_input = float(user_input)
    print(f"Converted to float: {float_input}")
except ValueError:
    print("Cannot convert to float.")

bool_input = bool(user_input)
print(f"Converted to boolean: {bool_input}")

# 10. Type casting with list elements: Convert each number in a list to a string.
my_list = [1, 2, 3]
str_list = [str(i) for i in my_list]
print("List of strings:", str_list)

# 11. Check if a number is positive, negative, or zero.
num = float(input("Enter a number: "))
if num > 0:
    print("The number is positive.")
elif num < 0:
    print("The number is negative.")
else:
    print("The number is zero.")

# 12. Print numbers from 1 to 10 using a for loop.
for i in range(1, 11):
    print(i)

# 13. Find the sum of all even numbers between 1 and 50.
sum_of_evens = sum(i for i in range(1, 51) if i % 2 == 0)
print("Sum of all even numbers between 1 and 50:", sum_of_evens)

# 14. Reverse a string entered by the user using a while loop.
input_str = input("Enter a string to reverse: ")
reversed_str = ""
i = len(input_str) - 1
while i >= 0:
    reversed_str += input_str[i]
    i -= 1
print("Reversed string:", reversed_str)

# 15. Calculate the factorial of a number entered by the user.
num = int(input("Enter a number to calculate its factorial: "))
factorial = 1
for i in range(1, num + 1):
    factorial *= i
print(f"The factorial of {num} is: {factorial}")
```
